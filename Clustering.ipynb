```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data 
data = pd.ExcelFile("PATIENT2.xlsx")
AAo = pd.read_excel(data,"AAo")
AArch = pd.read_excel(data,"AArch")

# Show dataframe
AAo
```

```python
AArch
```

```python
AAo = AAo.drop(["Diameter"], axis=1)
AAo = AAo.drop(["Kinetic Energy"], axis=1)
AAo
```

```python
AArch = AArch.drop(["Diameter"], axis=1)
AArch = AArch.drop(["Kinetic Energy"], axis=1)
AArch
```

```python
AAo = AAo.drop(AAo[AAo["Classification"]==3].index)
AAo
```

```python
AArch = AArch.drop(AArch[AArch["Classification"]==3].index)
AArch
```

```python
AAo = AAo.drop(["Classification"], axis=1)
AAo
```

```python
AArch = AArch.drop(["Classification"], axis=1)
AArch
```

```python
plt.figure(figsize=(15,10))
plt.title("AAo")
correlations = AAo.corr()
sns.heatmap(round(correlations,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
plt.figure(figsize=(15,10))
plt.title("AArch")
correlations = AArch.corr()
sns.heatmap(round(correlations,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform

plt.figure(figsize=(12,5))
plt.title("AAo")
dissimilarity = 1 - abs(correlations)
Z = linkage(squareform(dissimilarity), 'complete')

dendrogram(Z, labels=AAo.columns, orientation='top', 
           leaf_rotation=90);
```

```python
plt.figure(figsize=(12,5))
plt.title("AArch")
dissimilarity = 1 - abs(correlations)
Z2 = linkage(squareform(dissimilarity), 'complete')

dendrogram(Z2, labels=AArch.columns, orientation='top', 
           leaf_rotation=90);
```

```python
# Clusterize the data (AAo)
threshold = 0.8
labels = fcluster(Z, threshold, criterion='distance')

# Show the cluster
labels
```

```python
# Clusterize the data (AArch)
threshold = 0.8
labels2 = fcluster(Z2, threshold, criterion='distance')

# Show the cluster
labels2
```

```python
import numpy as np

# Keep the indices to sort labels
labels_order = np.argsort(labels)

# Build a new dataframe with the sorted columns
for idx, i in enumerate(AAo.columns[labels_order]):
    if idx == 0:
        clustered = pd.DataFrame(AAo[i])
    else:
        df_to_append = pd.DataFrame(AAo[i])
        clustered = pd.concat([clustered, df_to_append], axis=1)
```

```python
# Keep the indices to sort labels
labels_order = np.argsort(labels2)

# Build a new dataframe with the sorted columns
for idx, i in enumerate(AArch.columns[labels_order]):
    if idx == 0:
        clustered2 = pd.DataFrame(AArch[i])
    else:
        df_to_append = pd.DataFrame(AArch[i])
        clustered2 = pd.concat([clustered2, df_to_append], axis=1)
```

```python
plt.figure(figsize=(15,10))
plt.title("AAo")
correlations = clustered.corr()
sns.heatmap(round(correlations,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
plt.figure(figsize=(15,10))
plt.title("AArch")
correlations2 = clustered2.corr()
sns.heatmap(round(correlations2,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
sns.clustermap(correlations, method="complete", cmap='RdBu', annot=True, 
               annot_kws={"size": 7}, vmin=-1, vmax=1, figsize=(15,12));
```

```python
sns.clustermap(correlations2, method="complete", cmap='RdBu', annot=True, 
               annot_kws={"size": 7}, vmin=-1, vmax=1, figsize=(15,12));
```

```python
AAo_and_AArch = pd.concat([AAo, AArch], axis = 0)
AAo_and_AArch
```

```python
plt.figure(figsize=(15,10))
plt.title("AAo + AArch ")
correlations3 = AAo_and_AArch.corr()
sns.heatmap(round(correlations3,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
plt.figure(figsize=(12,5))
plt.title("AAo + AArch ")
dissimilarity3 = 1 - abs(correlations3)
Z3 = linkage(squareform(dissimilarity3), 'complete')

dendrogram(Z, labels=AAo_and_AArch.columns, orientation='top', 
           leaf_rotation=90);
```

```python
# Clusterize the data (AAo + AArch)
threshold = 0.8
labels3 = fcluster(Z3, threshold, criterion='distance')

# Show the cluster
labels3
```

```python
labels_order = np.argsort(labels3)

# Build a new dataframe with the sorted columns
for idx, i in enumerate(AAo_and_AArch.columns[labels_order]):
    if idx == 0:
        clustered3 = pd.DataFrame(AAo_and_AArch[i])
    else:
        df_to_append = pd.DataFrame(AAo_and_AArch[i])
        clustered3 = pd.concat([clustered3, df_to_append], axis=1)
```

```python
plt.figure(figsize=(15,10))
plt.title("AAo + AArch")
correlations3 = clustered3.corr()
sns.heatmap(round(correlations3,2), cmap='RdBu', annot=True, 
            annot_kws={"size": 7}, vmin=-1, vmax=1);
```

```python
sns.clustermap(correlations3, method="complete", cmap='RdBu', annot=True, 
               annot_kws={"size": 7}, vmin=-1, vmax=1, figsize=(15,12));
```
